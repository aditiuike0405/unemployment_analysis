```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
df=pd.read_csv("Unemployment in India.csv")
```

```python
df_cleaned = df.dropna(subset=['Longitude', 'Latitude'])
df=df_cleaned.copy()
```

```python
df[' Date'] = pd.to_datetime(df[' Date'].str.strip(), format='%d-%m-%Y')
```

```python
numeric_columns = [' Estimated Unemployment Rate (%)', ' Estimated Employed', ' Estimated Labour Participation Rate (%)']
correlation_matrix = df[numeric_columns].corr()
pastel_palette = sns.color_palette("pastel")
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap=pastel_palette, fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x=' Date', y=' Estimated Unemployment Rate (%)', hue='Region', data=df)
plt.title('Estimated Unemployment Rate Over Time by Region')
plt.xlabel('Date')
plt.ylabel('Estimated Unemployment Rate (%)')
plt.legend(title='Region', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
boxplot = sns.boxplot(x='Region', y=' Estimated Unemployment Rate (%)', data=df)
plt.title('Boxplot of Estimated Unemployment Rate by Region')
plt.xlabel('Region')
plt.ylabel('Estimated Unemployment Rate (%)')

boxplot.set_xticklabels(boxplot.get_xticklabels(), rotation=90)
plt.show()
```

```python
plt.figure(figsize=(6, 6))
sns.violinplot(x='Area', y=' Estimated Unemployment Rate (%)', data=df, inner='quartile')
plt.title('Distribution of Estimated Unemployment Rate by Area')
plt.xlabel('Area')
plt.ylabel('Estimated Unemployment Rate (%)')
plt.show()
```

```python
numeric_columns = [' Estimated Unemployment Rate (%)', ' Estimated Employed', ' Estimated Labour Participation Rate (%)']
pairplot = sns.pairplot(df[numeric_columns], kind='scatter')
pairplot.fig.suptitle('Scatter Plot Matrix for Selected Numeric Columns', y=1.02)
plt.show()
```

```python
# Geographical scatter plot of regions
plt.figure(figsize=(14, 10))
sns.scatterplot(x='Longitude', y='Latitude', hue='Region', size=' Estimated Employed', sizes=(20, 200), data=df, palette='Set2', alpha=0.7)
plt.title('Geographical Scatter Plot of Regions')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.legend(loc='upper right')
plt.show()
```
